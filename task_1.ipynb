**Часть 1**

```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None
```

```python
tests = ['12', ' -7 ', '12a', '']

for t in tests:
    print(f'{t!r} -> {parse_int(t)}')
```

**Часть 2**

```python
nums = [5, -3, 10, 0, 8, -7, 2, 14, -1, 6, 3, 12, -9, 4, 11, 20, -4, 7, 16, -2, 9]
print(f'кол-во элементов = {len(nums)}')
```

```python
def count_positive(nums: list[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count
```

```python
def sum_even(nums: list[int]) -> int:
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total
```

```python
def first_divisible(nums: list[int], k: int) -> int | None:
    for n in nums:
        if n % k == 0:
            return n
    return None
```

```python
print(f'Количество положительных: {count_positive(nums)}')
print(f'Сумма четных: {sum_even(nums)}')
print(f'Первое число, делящееся на 5: {first_divisible(nums, 5)}')
```

**Часть 3**

```python
items = [
    {'name': 'A', 'price': 120, 'rating': 4.7},
    {'name': 'B', 'price': 90, 'rating': 4.2},
    {'name': 'C', 'price': 150, 'rating': 4.9},
]
```

```python
sorted_by_price = sorted(items, key = lambda x: x['price'])
print('Сортировка по цене:')
print(sorted_by_price)
```

```python
sorted_by_rating = sorted(items, key = lambda x: x['rating'], reverse = True)
print('Сортировка по рейтингу:')
print(sorted_by_rating)
```

**Часть 4**

```python
class Chocolate:

    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError('Цена должна быть больше нуля')
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)
```

```python
choco1 = Chocolate('Milka', 120)
choco2 = Chocolate('kitkat', 200)

print(f'Цена со скидкой 10 процентов для Milka: {choco1.discount(10)}')
print(f'Цена со скидкой 15 процентов для kitkat: {choco2.discount(15)}')
```
